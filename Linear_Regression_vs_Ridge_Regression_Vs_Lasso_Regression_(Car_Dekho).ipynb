###Let's check distribution for cintinous variables.

```
# Checking distibution for all continous variables.
numeric_feature_list = car_df.describe().columns
numeric_feature_list
```

```
# Let's plot distplot for numeric variables to check the skewness.
 
for var in numeric_feature_list:
  sns.distplot(car_df[var], color="g")
  plt.title(f"Distribution of {var}")
  plt.xlabel(var)
  plt.show()
```

```
# Let's plot distplot for numeric variables with log transformation.
 
for var in numeric_feature_list:
  sns.distplot(np.log10(car_df[var]+1), color="g")
  plt.title(f"Distribution of {var}")
  plt.xlabel(var)
  plt.show()
```
